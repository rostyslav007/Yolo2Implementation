```python
import os
from glob import glob
import seaborn as sns
import random
import matplotlib.pyplot as plt
```

<h1 style='color: blue'>Anchor Boxes search</h1>

```python
images_path = glob(os.path.join('data', 'images', '*'))
```

```python
boxes_path = []
```

```python
for img_path in images_path:
    path = img_path.split('\\')
    name, extension = path[-1].split('.')
    boxes_path.append(os.path.join(*path[:-2], 'labels', name + '.txt'))
```

```python
train_part = 0.97
```

```python
train_boxes = boxes_path[:int(len(boxes_path)*train_part)]
```

```python
len(train_boxes)
```

```python
box_sizes = []
```

```python
for path in train_boxes:
    try:
        with open(path) as f:
            for line in f.readlines():
                w, h = line.split()[-2:]
                box_sizes.append((float(w), float(h)))
    except Exception as e:
        print(e)
```

```python
len(box_sizes)
```

```python
# Intersection over Union calculation
def iou(box1, box2):
    w1, h1 = box1
    x1, y1, x2, y2 = -w1/2, -h1/2, w1/2, h1/2
    
    w2, h2 = box2
    x3, y3, x4, y4 = -w2/2, -h2/2, w2/2, h2/2
    
    ix1 = max(x1, x3)
    iy1 = max(y1, y3)
    ix2 = min(x2, x4)
    iy2 = min(y2, y4)
    
    ih = max(iy2 - iy1, 0)
    iw = max(ix2 - ix1, 0)
    
    i_area = ih * iw
    
    # Area(A union B) = Area(A) + Area(B) - Area(A intersection B)
    
    u_area = h1*w1 + h2*w2 - i_area
    
    return i_area / u_area
```

```python
# distance function for 2 boxes
def distance(box1, box2):
    return 1 - iou(box1, box2)
```

```python
w_dimension = [b[0] for b in box_sizes]
h_dimension = [b[1] for b in box_sizes]
```

```python
# box shapes distribution
sns.scatterplot(x=w_dimension, y=h_dimension)
```

```python
k = 10
```

```python
# Custom K Means clustering algorithm
def k_means(points, centroids, epsilon = 1e-6):
    stop = False
    while not stop:
        color_dict = {i: [] for i in range(k)}
        
        for p in points:
            min_dist = float('inf')
            min_index = -1
            for i in range(k):
                dist = distance(centroids[i], p)
                if dist < min_dist:
                    min_dist = dist
                    min_index = i
            color_dict[min_index].append(p)
        
        new_centroids = []
        for key, value in color_dict.items():
            mean_w = sum([v[0] for v in value]) / len(value)
            mean_h = sum([v[1] for v in value]) / len(value)
            new_centroids.append((mean_w, mean_h))
        
        stop = all([(n_b[0] - n_b[0] < epsilon) and (n_b[1] - b[1] < epsilon) for (n_b, b) in zip(centroids, new_centroids)])
        
        centroids = new_centroids
    return centroids
```

```python
# starting centroids
start_centroids = random.sample(box_sizes, k)
```

```python
start_centroids
```

```python
centroids = k_means(box_sizes, start_centroids)
```

```python
centroids
```

```python
hue = []
for p in box_sizes:
    min_dist = float('inf')
    min_index = -1
    for i in range(k):
        dist = distance(centroids[i], p)
        if dist < min_dist:
            min_dist = dist
            min_index = i
    hue.append(min_index)
```

```python
ax = sns.scatterplot(x=w_dimension, y=h_dimension, hue=hue)
ax = sns.scatterplot(x=[c[0] for c in centroids], y=[c[1] for c in centroids], palette=['black']*k, \
                     hue=range(k), s=70, ec='black', legend=False, ax=ax)
plt.show()
```

```python
# saving anchors
f = open('data/anchors.txt', 'w')
for h, w in centroids:
    f.write(str(h) + ' ' + str(w) + '\n')
f.close()
```
